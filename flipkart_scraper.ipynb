```python
import pandas as pd
from selenium import webdriver
import time
from selenium.webdriver.common.keys import Keys
```

```python
products=['laptops', 'Phones', 'Headphones', 'smart watches', 'Cameras', 'Printers', 'monitors', 'Home theater', 'router']
```

```python
driver=webdriver.Chrome(r'Chromedriver.exe')
driver.get('https://www.flipkart.com/')
driver.find_element_by_xpath('//div[@class="_2QfC02"]/button').click()
driver.maximize_window()

flip_urls=[]
for product in products:
    driver.find_element_by_xpath('//input[@class="_3704LK"]').send_keys(product)
    driver.find_element_by_xpath('//button[@class="L0Z3Pu"]').click()
    time.sleep(2)
    j=0
    urls=[]
    url_tags=[]
    while j<11:
        time.sleep(2)
        try:
            url_tags=driver.find_elements_by_xpath('//a[@class="_1fQZEK"]')
        except:
            url_tags=driver.find_elements_by_xpath('//a[@class="_2rpwqI"]')
        for tag in url_tags:
            urls.append(tag.get_attribute('href'))
        try:
            next_button=driver.find_elements_by_xpath('//a[@class="_1LKTO3"]')[-1].click()
        except:pass
                
            
        j+=1
    flip_urls.append(urls)
    driver.find_element_by_xpath('//input[@class="_3704LK"]').send_keys(Keys.CONTROL,'a')
    time.sleep(1)
    driver.find_element_by_xpath('//input[@class="_3704LK"]').send_keys(Keys.BACKSPACE)
    driver.find_element_by_xpath('//input[@class="_3704LK"]').clear()
```

```python
len(flip_urls)
```

```python
urls=[]
for i in range(0,9):
    for url in flip_urls[i]:
        urls.append(url)
```

```python
len(urls)
```

```python
flip_stars=[]
flip_reviews=[]
driver=webdriver.Chrome(r'Chromedriver.exe')
driver.maximize_window()
for url in urls:
    driver.get(url)
    stars=[]
    reviews=[]        
    try:
        
        driver.find_element_by_xpath('//div[@class="_3UAT2v _16PBlm"]').click()
        time.sleep(2)
        try:
            star_tags=driver.find_elements_by_xpath('//div[contains(@class, "_3LWZlK")]')
            for tag in star_tags:
                stars.append(tag.text)
        except:pass
            #star.append('-')
        try:
            review_tags=driver.find_elements_by_xpath('//div[@class="t-ZTKy"]/div/div')
            for tag in review_tags:
                reviews.append(tag.text)
        except:pass
        for star in stars[1:]:
            flip_stars.append(star)
        for review in reviews:
            flip_reviews.append(review)
    except:
        pass
```
